```python
def test_a(A: int) -> bool:
    for x in range(0,100):
        for y in range(0,100):
            if not ((not(x<5) or (x**2<A)) and (not(y**2<=A) or (y<=5))):
                return False
    return True
```

```python
count = 0
for A in range(-100,100):
    if test_a(A):
        count += 1
print(count)
```

```python
def factorial(n: int) -> int:
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

```python
from functools import cache

@cache
def fib_with_cache(n: int) -> int:
    if n <= 1:
        return n
    return fib_with_cache(n-1)+fib_with_cache(n-2)

fib_with_cache(100)
```

```python
fib = [0] * 100
fib[0] = 0
fib[1] = 1

for n in range(2,100):
    fib[n] = fib[n-1] + fib[n-2]
    
fib[30]
```

```python
X = 1

#y = 2x + 1
def count_ways(y):
    #print(f'1 -> {y}')
    """   1  Y"""
    if y == X:
        return 1
    if y < 0 or y == 26:
        return 0
    if y % 2 == 1: # 1 -> 27 = 1 -> 26   +   1 -> 13
        return count_ways(y-1) + count_ways((y-1) // 2)
    else:
        return count_ways(y-1)
    
count_ways(27)
    
```

```python
import sys
import functools

sys.setrecursionlimit(2000000000)

history = {}
def f(n):
    if n == 0:
        return 0
    if n % 2 == 0:
        return f(n/2)
    if n-1 not in history:
        history[n-1] = f(n-1) + 1
    return history[n-1]

result = 0
for i in range(1000000000+1):
    if f(i) == 3:
        result += 1
result
```

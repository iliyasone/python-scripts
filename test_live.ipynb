```python
def transform(n: int) -> int:
    a = str(n)

    d1 = int(a[0]) + int(a[1])
    d2 = int(a[1]) + int(a[2])
    d3 = int(a[2]) + int(a[3])
    numbers = [d1, d2, d3]
    numbers.sort()
    x, y = numbers[1:]
    word = str(x) + str(y)
    
    return int(word)
```

```python
for i in range(9999, 999, -1):
    if i == transform(i):
        print(i, transform(i))

```

```python

def transofrm(n: int) -> int:
    b = bin(n)[2:]
    
    if n % 3 == 0:
        b += b[-3:]
    else:
        b += bin((n % 3) * 3)[2:]
        
    return int(b,2)

transofrm(4)
```

```python
p = max_ = 0
for i in range(0, 10000):
    r = transofrm(i)
    if r > max_ and r < 170:
        max_ = r
        p = i
        print(max_, p)
max_, p
        
```
